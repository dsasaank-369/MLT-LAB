```python
import io
import sys
import pandas as pd
```

```python
df2 = pd.read_csv(r"123014019/iris.data")
```

```python
print((df2.shape))
print(len(df2.columns))
print(df2.head)
print(df2.dtypes)
```

```python
hist = df2.hist()
```

```python
ax1 = df2.plot.scatter(x='sepal length',y='sepal width',c='DarkRed',title='sepal length vs sepal width')
```

```python
import seaborn as sbn
df = df2.groupby('class')['petal length'].agg(['mean', 'min', 'max'])
ax = sbn.lineplot(data=df, palette="bright")
ax.set(xlabel='class', ylabel='petal length')
```

```python
df = df2.groupby('class')['petal width'].agg(['mean', 'min', 'max'])
ax = sbn.lineplot(data=df, palette="dark")
ax.set(xlabel='class', ylabel='petal width')
```

```python
df = df2.groupby('sepal width')['petal width'].agg(['mean', 'min', 'max'])
ax = sbn.lineplot(data=df, palette="dark")
ax.set(xlabel='class', ylabel='petal width')
```

```python
df = df2.groupby('petal length')['petal width'].agg(['mean', 'min', 'max'])
ax = sbn.lineplot(data=df, palette="dark")
ax.set(xlabel='class', ylabel='petal width')
```

```python
def best_petal(row):
  if(row['petal length']>1.5 and row['class']=="Iris-virginica"):
    return 0
  else:
    return 1
```

```python
df2['best_petal'] = df2.apply (lambda row: best_petal(row), axis=1)
print(df2)
```
