```python
import matplotlib.pyplot as plt

from sklearn import datasets

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
wine = datasets.load_wine()

X = wine.data
y = wine.target
target_names = wine.target_names
```

```python
lda = LDA(n_components=2)
lda_X = lda.fit(X,y).transform(X)
```

```python
plt.scatter(lda_X[y == 0, 0], lda_X[y == 0, 1], s =50, c = 'orange', label = 'Ash')
plt.scatter(lda_X[y == 1, 0], lda_X[y == 1, 1], s =50,  c = 'yellow', label = 'Alcohol')
plt.scatter(lda_X[y == 2, 0], lda_X[y == 2, 1], s =50,  c = 'green', label = 'Magnesium')
plt.title('LDA plot for Wine Dataset')
```
